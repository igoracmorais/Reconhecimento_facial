```python
## Packages
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Flatten
from keras.layers import Convolution2D, MaxPooling2D
from keras.utils import np_utils
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
diretorio = os.getcwd()
```

## 1. Simple deep learning model

```python
dataset=np.loadtxt(diretorio+"/"+"pima-indians-diabetes.csv",delimiter=",")
X=dataset[:,0:8]
Y=dataset[:,8]
```

### 2. Define model

```python
model=Sequential()
model.add(Dense(12,input_dim=8,activation="relu"))  #first layer with inputs = 8 and 12 neurons
model.add(Dense(8,activation="relu"))    #second layer with 8 neurons
model.add(Dense(1,activation="sigmoid"))   #layer de output
```

### 3. Compile model

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

### 4. Fit model

```python
model.fit(X,Y,epochs=150,batch_size=10)   #use verbose=0 to not print results
```

### 5. Evaluate model

```python
scores=model.evaluate(X,Y)
print("\n %s: %.2f%%" % (model.metrics_names[1],scores[1]*100))
```

```python
scores
```

### 6. forecasting

```python
prev=model.predict(X)
prev
```

```python
prev2=[round(x[0]) for x in prev]
prev2
```

## 2. Multilayer Perceptron (MLP)

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD

# Generate dummy data
import numpy as np
x_train = np.random.random((1000, 20))
y_train = keras.utils.to_categorical(np.random.randint(10, size=(1000, 1)), num_classes=10)
x_test = np.random.random((100, 20))
y_test = keras.utils.to_categorical(np.random.randint(10, size=(100, 1)), num_classes=10)
```

```python
y_train[65]
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=20))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))   #number of output dimension

sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=20,batch_size=128,verbose=0)
score = model.evaluate(x_test, y_test, batch_size=128)
```

```python
score[1]*100
```

```python
previs=model.predict(x_test)
plt.plot(previs[2])
plt.plot(previs[4])
plt.plot(previs[10]);
```
